## 2. Untuned Benchmark

```python
X_train = df[df.date_block_num < 33].drop(['item_cnt_month','item_category_name', 'group_name'], axis=1)
y_train = df[df.date_block_num < 33]['item_cnt_month']
X_valid = df[df.date_block_num == 33].drop(['item_cnt_month','item_category_name', 'group_name'], axis=1)
y_valid = df[df.date_block_num == 33]['item_cnt_month']
X_test = df[df.date_block_num == 34].drop(['item_cnt_month','item_category_name', 'group_name'], axis=1)
```

```python
model = xgb.XGBRegressor(tree_method='gpu_hist') #if you don't have a gpu and want to run, simply remove this parameter

ts = time.time()

model.fit(X_train, y_train, eval_metric="rmse", eval_set=[(X_train, y_train), (X_valid, y_valid)], verbose=True)

print(time.time()-ts, "seconds", (time.time()-ts)/60.0,"minutes")
```

```python
#Does rounding help? 
y_valid_pred = model.predict(X_valid).clip(0,20)
y_valid_pred_rounded = [round(value) for value in y_valid_pred]
```

```python
from sklearn.metrics import accuracy_score, r2_score, mean_squared_error

r2 = r2_score(y_valid, y_valid_pred)
print("R-squared score:",(r2))
r2 = r2_score(y_valid, y_valid_pred_rounded)
print("R-squared score (rounded prediction):",(r2))

print("")

mse = mean_squared_error(y_valid, y_valid_pred)
print("Mean Squared Error:",(mse))
mse = mean_squared_error(y_valid, y_valid_pred_rounded)
print("Mean Squared Error (rounded prediction):",(mse))

print("")

rmse = mean_squared_error(y_valid, y_valid_pred, squared=False)
print("Root mean Squared Error:",(rmse))
rmse = mean_squared_error(y_valid, y_valid_pred_rounded, squared=False)
print("Root mean Squared Error (rounded prediction):",(rmse))
```

The rounded values are better for R-squared but the RMSE is higher, so we will not round. 

```python
plt.rcParams["figure.figsize"] = (16, 8)
xgb.plot_importance(model, max_num_features=12)
```

Here we see the most important feaures: revenue of an item, the item monthly average price, and which shop and item ID they have. Seeing these features as important is intuitive! 

```python
y_test = model.predict(X_test).clip(0,20)
# y_test = [round(value) for value in y_test]
```

## 3. Other algorithms

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold


from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# Now that we are introducing cross fold validation and we are not validating, I am going to include the last month. 
X_train = df[df.date_block_num < 34].drop(['item_cnt_month','item_category_name', 'group_name'], axis=1)
y_train = df[df.date_block_num < 34]['item_cnt_month']
X_test = df[df.date_block_num == 34].drop(['item_cnt_month','item_category_name', 'group_name'], axis=1)
```

```python
pipelines = []
pipelines.append(('UnscaledLR', Pipeline([('LR',LinearRegression())])))
pipelines.append(('ScaledLR', Pipeline([('Scaler', StandardScaler()),('LR',LinearRegression())])))
pipelines.append(('UnscaledLASSO', Pipeline([('LASSO', Lasso())])))
pipelines.append(('ScaledLASSO', Pipeline([('Scaler', StandardScaler()),('LASSO', Lasso())])))
pipelines.append(('UnscaledEN', Pipeline([(('EN', ElasticNet()))])))
pipelines.append(('ScaledEN', Pipeline([('Scaler', StandardScaler()),('EN', ElasticNet())])))
pipelines.append(('UnscaledCART', Pipeline([('CART', DecisionTreeRegressor())])))
pipelines.append(('ScaledCART', Pipeline([('Scaler', StandardScaler()),('CART', DecisionTreeRegressor())])))
pipelines.append(('UnscaledXGB', Pipeline([('XGB', xgb.XGBRegressor(tree_method='gpu_hist'))])))
pipelines.append(('ScaledXGB', Pipeline([('Scaler', StandardScaler()),('XGB', xgb.XGBRegressor(tree_method='gpu_hist'))])))
```

```python
results = []
names = []
for name, model in pipelines:
    ts = time.time()
    kfold = KFold(n_splits=5)#, random_state=42
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring='neg_root_mean_squared_error')
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
    print(time.time()-ts, "seconds", (time.time()-ts)/60.0,"minutes")
```

We can see the two best scores here are CART (Decision Tree Regression) and XGB (Extreme Gradient Boosting Regression); it seems like scaling doesn't make a significant difference so I will not be scaling the data. 

# 4. Hyperparameter Tuning

Here I am going to use cross validation and grid search to find the best hyperparameters for the two top scoring algorithms. 

```python
# Now that we are introducing cross fold validation and we are not validating, I am going to include the last month. 
X_train = df[df.date_block_num < 34].drop(['item_cnt_month','item_category_name', 'group_name'], axis=1)
y_train = df[df.date_block_num < 34]['item_cnt_month']
X_test = df[df.date_block_num == 34].drop(['item_cnt_month','item_category_name', 'group_name'], axis=1)
```

```python
# Make the models 
dt = DecisionTreeRegressor()
xgbr = xgb.XGBRegressor(tree_method='gpu_hist')
```

```python
print(dt.get_params())
print(xgbr.get_params())
```

```python
# Import GridSearchCV
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
```

### XGB model

```python
# Define the grid of hyperparameters 'params_xgbr'
params_xgbr = {
'n_estimators': [200, 300, 400],
'max_depth': [3, 4, 5, 6],
'min_samples_leaf': [0.01, 0.05, 0.1],
'max_features': ['log2', 'sqrt'], 
'tree_method': ['gpu_hist'] #I wanted to make sure we were on my GPU by inclduing this here as well as in the initial declaration 
}
```

```python
# Instantiate a 5-fold CV grid search object
tscv = TimeSeriesSplit() # Defaults to 5 splits

grid_xgbr = GridSearchCV(estimator=xgbr,
param_grid=params_xgbr,
scoring='neg_root_mean_squared_error',
verbose=1,
cv=tscv)
```

```python
# Fit 'grid_dt' to the training data
import warnings
warnings.filterwarnings('ignore')

ts = time.time()

grid_xgbr.fit(X_train, y_train)

print(time.time()-ts, "seconds", (time.time()-ts)/60.0,"minutes")
```

```python
# Extract best hyperparameters from 'grid_xgbr'
best_hyperparams = grid_xgbr.best_params_
print('Best hyerparameters:\n', best_hyperparams)
```

```python
from sklearn.metrics import mean_squared_error as MSE
# Extract best model from 'grid_xgbr'
best_model_xgb = grid_xgbr.best_estimator_
```

```python
# Extract best CV score from 'grid_xgbr'
best_CV_score = grid_xgbr.best_score_
print('Best CV RMSE: {:.3}'.format(best_CV_score))
```

```python
pred_XGB = best_model_xgb.predict(X_test).clip(0,20)
```

```python
pred_XGB = pd.DataFrame({"ID": test.index, "item_cnt_month": pred_XGB})
datapath = './data/modeling'
save_file(pred_XGB, 'tuned_XGB.csv', datapath)
```

### Decision Tree (CART) model

```python
# Define the grid of hyperparameters 'params_dt'
params_dt = {
'max_depth': [3, 4, 5, 6],
'min_samples_leaf': [0.04, 0.06, 0.08],
'max_features': [0.2, 0.4, 0.6, 0.8]
}
```

```python
# Instantiate a 5-fold CV grid search object 'grid_dt'
grid_dt = GridSearchCV(estimator=dt,
param_grid=params_dt,
scoring='neg_root_mean_squared_error',
cv=tscv)
```

```python
import warnings
warnings.filterwarnings('ignore')

ts=time.time()
# Fit 'grid_dt' to the training data
grid_dt.fit(X_train, y_train)

print(time.time()-ts, "seconds", (time.time()-ts)/60.0,"minutes")
```

```python
# Extract best hyperparameters from 'grid_dt'
best_hyperparams = grid_dt.best_params_
print('Best hyerparameters:\n', best_hyperparams)
```

```python
# Extract best CV score from 'grid_dt'
best_CV_score = grid_dt.best_score_
print('Best CV RMSE: {:.3}'.format(best_CV_score))
```

```python
# Extract best model from 'grid_dt'
best_model_dt = grid_dt.best_estimator_
```

```python
pred_CART = best_model_dt.predict(X_test).clip(0,20)
```

```python
pred_CART.shape
```

```python
tuned_CART = pd.DataFrame({"ID": test.index, "item_cnt_month": pred_CART})
datapath = './data/modeling'
save_file(tuned_CART, 'tuned_CART.csv', datapath)
```

### Average the other two predictions

```python
pred_cart_noclip = best_model_dt.predict(X_test)
```

```python
pred_xgb_noclip = best_model_xgb.predict(X_test)
```

```python
average = (pred_cart_noclip + pred_xgb_noclip) /2.0
```

```python
average = average.clip(0,20)
```

```python
tuned_average = pd.DataFrame({"ID": test.index, "item_cnt_month": average})
datapath = './data/modeling'
save_file(tuned_average, 'tuned_average.csv', datapath)
```
