## 2. Is there a relationship between certain emotions and the rater’s confidence in those emotions? (i.e. are some emotions inherently less obvious to raters?) Make you case both statistically and with visualizations

First, I will perofrm a one way ANOVA on the data. 

```python
df_emotions = df[['emotion','emotion:confidence']]
df_emotions = df_emotions.rename(columns={"emotion": "emotion", "emotion:confidence": "confidence"})
```

```python
model = ols('confidence ~ C(emotion)', data=df_emotions).fit()
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```

The PR(>F) is incredibly small, showing there is a statistically significant correlation between emotion and confidence in the emotion. 

Here it is with scipy:

```python
# f_oneway() function takes the group data as input and 
# returns F-statistic and P-value
from scipy.stats import f_oneway
 
# Assumption(H0) is that FuelType and CarPrices are NOT correlated
 
# Finds out the Prices data for each FuelType as a list
CategoryGroupLists=df.groupby('emotion')['emotion:confidence'].apply(list)
 
# Performing the ANOVA test
# We accept the Assumption(H0) only when P-Value <= 0.05 
AnovaResults = f_oneway(*CategoryGroupLists)
print('P-Value for Anova is: ', AnovaResults[1])
```

PR(>F) is the same as the ANOVA; there is a correlation. 

Now I will show any correlation graphically. 

```python
encodings = pd.get_dummies(df['emotion'], prefix='emo')
df = pd.concat([df, encodings], axis=1)
df.sample(5)
```

## 3. Build a python (object oriented) class that is designed to tokenize the sentences and keep track of which emotions get associated with each word.
    A. Should have a method add_one_sentence(sentence, emotion) that breaks the sentence into words (i.e. unigrams) that are case insensitive and logs counts of each word-emotion association
    B. Should have a function to return the frequency of each emotion for an individual word normalized by the base rates for the emotions (i.e. how frequently is a specific categorized as each emotion, after accounting for how frequent each emotion is in the dataset)
    C. Plot these distributions for the words that 1) appear more than 100 times but 2) have the emotion distributions that are the most different from the base rates of each emotion (based on whatever metric you like)

```python
stop_words = set(stopwords.words('english'))
```

```python
class Emotion_Tracker:
    # attributes
    def __init__(self, name):
        self.name = name
        self.emo_dict = {}
        self.emotions = Counter()
        self.total_words = Counter()
        
    # add_one_sentence method
    # takes in sentence, emotion
    # breaks the sentence into words (unigrams) that are case insensitive 
    # logs counts of each word-emotion association
    def add_one_sentence(self, sentence, emotion): 
        tokens = nltk.word_tokenize(sentence) # tokenize the sentence 
        tokens = [word.lower() for word in tokens if word.isalpha()]
        filtered_tokens = [t.lower() for t in tokens if not t.lower() in stop_words]
        
        words = Counter(filtered_tokens)
        
        if emotion in self.emo_dict: 
            self.emo_dict[emotion] += words
        else:
            self.emo_dict[emotion] = words
            
        self.emotions.update({emotion})
            
    # returns the dictionary that logs emotions and the counters of words for that emotion
    def get_log(self): 
        return(self.emo_dict)
    
    # return the frequency of each emotion for an individual word normalized by the base rates for the emotions
    def word_frequency(self, word, emotion): 
        word_in_emotion_count = self.emo_dict[emotion][word]
        total_emotion_count = self.emotions[emotion]
        return word_in_emotion_count/total_emotion_count
    
    # function to update total word counter
    def update_total_words_count(self): 
        for emotion in df.emotion.unique():
            if emotion in self.emo_dict.keys(): 
                self.total_words.update(self.emo_dict[emotion])
                
    # function to find of the word is rare
    def rare(self, words):
        rare_words = []
        for word in words: 
            for emotion in self.emotions.keys():
                freq = self.word_frequency(word, emotion)
                if freq > 0.1: # TODO: make this a relative number, not a fixed 0.1
                    rare_words.append((word, freq, emotion))
        return(rare_words)

    # Plot these distributions for the words that 
    # 1) appear more than 100 times but 
    # 2) have the emotion distributions that are the most different from 
    # the base rates of each emotion (based on whatever metric you like)
    def plot(self):
        self.update_total_words_count()
        
        words_100 = [w for w, v in self.total_words.items() if v > 1] # words that appear more than 100 times
        rare_words = self.rare(words_100)
        #emotion in self.emotions.keys()
        rare_df = pd.DataFrame.from_records(rare_words, columns =['word', 'freq', 'emotion'])
        
        sns.barplot(x=rare_df.word, y=rare_df.freq, hue=rare_df.emotion)
    
```

```python
alpha = Emotion_Tracker('alpha')
```

```python
for ind in df.index:
    alpha.add_one_sentence(df['sentence'][ind], df['emotion'][ind])

alpha.plot()
```

## 4. Are the sentence vectors useful in predicting the emotion labels (1024 dim vectors in the last column)? Make your case either way. 
1.	In addition to statistical evidence, show some data visualization to show how the vector does or doesn’t map to emotion labels. 

Vectors of sentences are very useful in predicting label. Vectorization is the current best way to get from text data to something machine learning algorithms can use. There is some correlation between them just at a glance, but that isn't real data. 

```python
df.vector.values[0][0], df.vector.values[1][0]
```

```python
df.vector.values[50][0], df.vector.values[40][0]
```

Initially we see that they have similar values in the 0th index for the same emotions - in this case Neutral and Anger. But that's not proof. 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(list(df.vector.values), list(df.emotion.values), test_size=0.2, random_state = 1234)
```

```python
# Bernoulli Naïve Bayes model.
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline

p = Pipeline([('Normalizing',MinMaxScaler()),('MultinomialNB',MultinomialNB())])
p.fit(X_train,y_train) 

# naive_bayes = MultinomialNB()
# naive_bayes.fit(X_train, y_train)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
# evaluate on the test data
# make predictions on the test data
pred = p.predict(X_test)

print('accuracy score: ', accuracy_score(y_test, pred))
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

p = Pipeline([('Normalizing',MinMaxScaler()),('LogisticRegression',RandomForestClassifier())])
p.fit(X_train,y_train) 

# evaluate on the test data
# make predictions on the test data
pred = p.predict(X_test)

print('accuracy score: ', accuracy_score(y_test, pred))
```

```python
# logistic regression. 
from sklearn.linear_model import LogisticRegression

p = Pipeline([('Normalizing',MinMaxScaler()),('LogisticRegression',LogisticRegression())])
p.fit(X_train,y_train) 

# evaluate on the test data
# make predictions on the test data
pred = p.predict(X_test)

print('accuracy score: ', accuracy_score(y_test, pred))
```

```python
# Try neural network
from sklearn.neural_network import MLPClassifier

p = Pipeline([('Normalizing',MinMaxScaler()),('LogisticRegression',MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(21, 18), random_state=1234))])
p.fit(X_train,y_train) 

# evaluate on the test data
# make predictions on the test data
pred = p.predict(X_test)

print('accuracy score: ', accuracy_score(y_test, pred))
```

As seen by these machine learning algorithms, just throwing in these vectors into various ML algorithms doesn't prodouce meaningful results.
