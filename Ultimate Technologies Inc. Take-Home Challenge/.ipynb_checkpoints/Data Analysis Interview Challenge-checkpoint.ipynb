```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

The data ranges from Jan 1 1970 to April 13, 1970. These 4.5 months have 0 missing values. 

```python
logins['count'] = 1
logins.head()
```

### Group the data

```python
logins2 = logins
```

```python
logins2['count']=1
logins2.index = logins2["login_time"]
```

```python
logins2.head()
```

```python
fifteen = logins2.resample('15min')['count'].sum()
```

```python
fifteen.head()
```

```python
# Use pd.Grouper to group by 15 min intervals
login_grouped = pd.DataFrame(logins.groupby([pd.Grouper(key='login_time', freq='15Min')]).sum()).reset_index()
login_grouped.head()
```

```python
login_grouped.tail()
```

Now that the data is grouped into 15 minute intervals, let's examine the data. 

I want to look at time of day, day of week, and monthly views, as well as overall trends. 

```python
login_grouped['hour'] = login_grouped['login_time'].dt.hour
login_grouped['day_of_week'] = login_grouped['login_time'].dt.dayofweek
login_grouped['month'] = login_grouped['login_time'].dt.month
```

### 1.1: Time of day

```python
login_hourly = login_grouped.groupby('hour')['count'].sum()
```

```python
sns.set_style('whitegrid')
sns.set_context('poster')
palette=sns.color_palette("crest", n_colors=len(login_hourly.index))

plt.figure(figsize=(16, 8))
hourly_bar = sns.barplot(x=login_hourly.index, y=login_hourly, palette=palette)
hourly_line = sns.lineplot(x=login_hourly.index, y=login_hourly)
hourly_line.set_xticks(np.arange(0,25,1));
```

Most logins come during 11am-1pm, and 9pm-4am. 

### 1.2: Day of week

```python
login_day = login_grouped.groupby('day_of_week')['count'].sum()
login_day.head(10)
```

```python
sns.set_style('whitegrid')
sns.set_context('poster')
palette=sns.color_palette("crest", n_colors=len(login_day.index))


plt.figure(figsize=(16, 8))
hourly = sns.barplot(x=login_day.index, y=login_day, palette=palette)
hourly.set_xticks(np.arange(0,7,1));
hourly.set_xticklabels(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']);
```

Weekends get a lot more activity! 

### 1.2: Month of year

```python
login_month = login_grouped.groupby('month')['count'].sum()
```

```python
sns.set_style('whitegrid')
sns.set_context('poster')
palette=sns.color_palette("crest", n_colors=len(login_month.index))


plt.figure(figsize=(16, 8))
hourly = sns.barplot(x=login_month.index, y=login_month, palette=palette)
hourly.set_xticks(np.arange(0,4,1));
hourly.set_xticklabels(['Jan','Feb','Mar','Apr'])
```

### ARIMA summary

```python
import statsmodels.api as sm
from pylab import rcParams

y = fifteen.resample('d').sum()

y.dtypes
```

```python
rcParams['figure.figsize'] = 18, 12
decomposition = sm.tsa.seasonal_decompose(y, model='additive')
fig = decomposition.plot()
plt.show()
```

In this graph we confirm a few things we've already seen. The daily log in count varies a lot but we see a pattern to the log in times. The Trend graph shows a general rise to march then fall toward april, as we saw above. The Seasonal chart shows that every week, we peak logins consistently - those are the weekends we saw above.

## Part 2: Experiment and metrics design

The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on
weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active
during the day. On weekends, there is reasonable activity in both cities.

However, a toll bridge, with a two way toll, between the two cities causes driver partners to tend
to be exclusive to each city. The Ultimate managers of city operations for the two cities have

1. What would you choose as the key measure of success of this experiment in

2. Describe a practical experiment you would design to compare the effectiveness of the

    a) how you will implement the experiment
    
    b) what statistical test(s) you will conduct to verify the significance of the

    c) how you would interpret the results and provide recommendations to the city
operations team along with any caveats.

Note: The two cities of Gotham and Metropolis are not in the provided dataset; however, you do
not need this information to answer Part 2.

### 2.1: Choosing a metric

We are trying to encouraging driver partners to serve both cities. As such, we want to see more trips between the cities. The most obvious metric of number of reinbursements is not a good metric, as it does not show any increase to the current number of inter-city trips and many could sign up but not use the service - we cannot gauge if it is worth the cost of implementing. We could implement a new feature, called "number of inter-city trips", but this is not specific enough to our goal. 

A feature we already have is toll costs. An increase in this is correlated with an increase in new inter-city trips from driver partners. We can measure the toll costs reimbursed to partner drivers as well. Dividing the reimbursed cost by the total toll costs gives us a percentage of tolls that are from driver partners due to this initiative. The higher this number, the more partner drivers are actively using this reimbursement system. 

### 2.2: Practical experiment

First, I would colect data: the toll costs reimbursed, include date, time, cost, direction, driver parterns from which city, and toll costs incharged. Then I can aggregate and anlalyze the data to find out the trend. If the costs reimbursed and percentage are increasing (statistically significantly), then the imlpementation is successful. The results can be visualized through trend data, similar to Part 1. I would show day of the week, time, direction, and percentage of reimbursed toll cost relative to total toll costs) across time. The reccomendations would be relative to that data, i.e. a weekend special if weekend data is different to weekday data.

There are definintely a lot more high ratings than low ratings in these two columns. Therefore, it is likely that the low values are missing because of inherently low satisfaction! Filling in with the average rating would screw up the data.

```python
print("We are missing {} percent of driver rating data.".format(ultimate.avg_rating_of_driver.isna().sum() / len(ultimate.index) *100))
print("We are missing {} percent of rating by driver data.".format(ultimate.avg_rating_by_driver.isna().sum() / len(ultimate.index) *100))
```

## 3.2 Modeling

```python
ultimate.info()
```

```python
ultimate.head()
```

```python
# I want to retain the last trip date and signup_date data without keeping datetime columns, so I will make a feature "days since last trip"
# The refernce point in time with be 2014-07-01
ultimate['days_since_last_trip'] = -1
ultimate['days_since_signup'] = -1

ultimate['days_since_last_trip'] = (pd.to_datetime("2014-07-01") - pd.to_datetime(ultimate['last_trip_date'])).dt.days
ultimate['days_since_signup'] = (pd.to_datetime("2014-07-01") - pd.to_datetime(ultimate['signup_date'])).dt.days
ultimate.head()
```

```python
# Drop datetime (object) features
ultimate.drop(['signup_date', 'last_trip_date'], axis=1, inplace=True)
```

```python
# Creating dummy variables for categorical features
df = pd.get_dummies(ultimate, columns = ['city', 'phone', 'ultimate_black_user'], drop_first=True)
```

```python
df.head()
```

### 3.2.2 Modeling

```python
X = df.drop(['retained','days_since_last_trip'], axis=1)
y = df['retained']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# It is not difficult to test a variety of models. 
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

import time
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import xgboost as xgb
```

```python
pipelines = []
pipelines.append(('LR', Pipeline([('LR',LogisticRegression())]))) # Logistic Regression
pipelines.append(('NB', Pipeline([('NB', GaussianNB())]))) #Naive Bayes
pipelines.append(('SGD', Pipeline([(('SGD', SGDClassifier()))]))) # Stochastic Gradient Descent
pipelines.append(('KNN', Pipeline([('KNN', KNeighborsRegressor())]))) # K Nearest Neighbors
pipelines.append(('RF', Pipeline([('RF', RandomForestClassifier())]))) # Random Forest Classifier
pipelines.append(('DT', Pipeline([('DT', DecisionTreeClassifier())]))) # Decision Tree
# pipelines.append(('SVC', Pipeline([('SVC', SVC())]))) # Support Vector Machine
pipelines.append(('XGB', Pipeline([('XGB', xgb.XGBRegressor(tree_method='gpu_hist'))]))) # Gradient-boosted Decision Tree
```

```python
results = []
names = []
for name, model in pipelines:
    ts = time.time()
    kfold = KFold(n_splits=5)
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold) #scoring='accuracy'
    results.append(cv_results)
    names.append(name)
    msg = "{}: mean: {:.4f} std: {:.4f}".format(name, cv_results.mean(), cv_results.std())
    print(msg)
    print("{:.2f} seconds".format(time.time()-ts))
```

Our highest accuracy model is the Random Forest model, with an initial accuract of 77 percent. 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

RF = RandomForestClassifier()
cv_scores = cross_val_score(RF, X_train, y_train, cv=5)
print(cv_scores)
print("Average 5-Fold CV Score: {}".format(np.mean(cv_scores)))
```

```python
# Create the hyperparameter grid
n_estimators = [10,20,50,100,200]
max_features = [5, 8, 10, 'auto', None]
min_samples_split = [2, 4, 6]
param_grid = {'n_estimators': n_estimators, 'max_features': max_features, 'min_samples_split': min_samples_split}
```

```python
# Instantiate the GridSearchCV object
RF_cv = GridSearchCV(RF, param_grid, cv=5)
# build model    
RF_cv.fit(X_train, y_train)
# predict using model
y_predict = RF_cv.predict(X_test) 
```

```python
# Print the optimal parameters and best score
print("Tuned Parameter: {}".format(RF_cv.best_params_))
print("Tuned Score: {}".format(RF_cv.best_score_))
print()
# evaluate model prediction performance 
print('Test set performance:')
print ('Accuracy: ', accuracy_score(y_test, y_predict))
print (classification_report(y_test, y_predict))
```

### Final Model

```python
rfc = RandomForestClassifier(max_features='auto', min_samples_split=6, n_estimators=200)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
# Generate classification report
print(classification_report(y_test, y_pred))
print('Test set accuracy: {:.2f}%'.format(accuracy_score(y_test, y_pred)*100))
```

```python
# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
pd.DataFrame(cm, index=range(0,2), columns=range(0,2))
```

```python
# Compute predicted probabilities
y_pred_prob = rfc.predict_proba(X_test)[:,1]

# Generate ROC curve values: fpr, tpr, thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

# Plot ROC curve
plt.figure(figsize=(6,6))
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
```

```python
# Compute and print AUC score
print("AUC: {:.4f}".format(roc_auc_score(y_test, y_pred_prob)))
```

The final metrics of our model are: 

1. Test set accuracy: 77.38%
2. AUC: 0.8408

```python
feature_importance = pd.DataFrame(list(zip(X.columns, rfc.feature_importances_)), columns = ['features', 'Importance'])
feature_importance.sort_values(by='Importance', ascending=False).head(5)
```

From the importance of features, we find that the most important 3 features are:

- avg_dist: the average distance in miles per trip taken in the first 30 days after signup
- weekday_pct: the percent of the user’s trips occurring during a weekday
- avg_rating_by_driver: the rider’s average rating over all of their trips
